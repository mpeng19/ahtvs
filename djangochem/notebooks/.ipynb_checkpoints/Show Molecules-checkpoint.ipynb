```python
import os
import sys
import django
import pprint
import pandas as pd
sys.path
os.environ["DJANGO_SETTINGS_MODULE"] = "djangochem.settings.default"
django.setup()
from pgmols.models import *

from IPython.display import HTML
import matplotlib
%matplotlib inline
from rdkit.Chem import AllChem as Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from rdkit.Chem import PandasTools
PandasTools.molRepresentation = 'svg'
```

```python
def show(df):
    return HTML(df.to_html(escape=False))
```

```python
ms = [m.to_dict() for m in Mol.objects.all()]


Draw.MolsToGridImage([m['mol'] for m in ms], molsPerRow=3)
```

```python
df = pd.DataFrame(ms)
cols = ['mol', 'mass', 'stoichiometry']
df = df[cols]
show(df)
```
